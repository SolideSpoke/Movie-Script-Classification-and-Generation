```python
def clean_script(text):
    """
    Cleans up the script text by removing unnecessary whitespace and
    formatting issues. It replaces the "NEWLINE_TOKEN" with actual newlines
    and removes extra spaces before punctuation marks.
    It also removes leading and trailing whitespace from the entire text.
    Args:
        text (str): The script text to be cleaned.
    Returns:
        str: The cleaned script text.
    """
    import re

    text = text.replace(" NEWLINE_TOKEN ", "\n")
    text = text.replace(" NEWLINE_TOKEN", "\n")
    text = text.replace("NEWLINE_TOKEN ", "\n")

    text = re.sub(r"\s+([.,!?;:])", r"\1", text)

    return text.strip()
```

```python
import csv

file = "5-gram-50-tokens.csv"

csv_reader = csv.reader(open(file, "r"), delimiter=";")



for row in csv_reader:
    print(len(row))
    if len(row) == 9:
        seed, generate, reference, bleu, rouge, perplexity, f1, precision, recall = row
        print("=" * 20)
        print(f"Seed:\n{clean_script(seed)}")
        print("-" * 20)
        print(f"Generate:\n{clean_script(generate)}")
        print("-" * 20)
        print(f"Reference:\n{clean_script(reference)}")
        print("-" * 20)
        print(f"BLEU: {bleu}")
        print(f"ROUGE: {rouge}")
        print(f"Perplexity: {perplexity}")
        print(f"F1: {f1}")
        print(f"Precision: {precision}")
        print(f"Recall: {recall}")
    elif len(row) == 8:
        seed, generate, reference, bleu, rouge, f1, precision, recall = row
        print("=" * 20)
        print(f"Seed:\n{clean_script(seed)}")
        print("-" * 20)
        print(f"Generate:\n{clean_script(generate)}")
        print("-" * 20)
        print(f"Reference:\n{clean_script(reference)}")
        print("-" * 20)
        print(f"BLEU: {bleu}")
        print(f"ROUGE: {rouge}")
        print(f"F1: {f1}")
        print(f"Precision: {precision}")
        print(f"Recall: {recall}")
```
