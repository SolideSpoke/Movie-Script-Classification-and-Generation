```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/MovieDataThread.csv")
df.head()
```

```python
index_drop = df.loc[df['imdb_id'].isnull()].index
df = df.drop(index=index_drop)
df = df.drop(columns=['imdb_id'])
df_filtered = pd.concat([df[['Script','Title']],df.filter(like='imdb_')],axis=1)
```

```python
print(df_filtered.shape)
print(df_filtered.columns)
df_filtered.head()
```

```python
genre_counts = df.filter(like='imdb_').count().sort_values(ascending=False)
print(genre_counts)

plt.figure(figsize=(12, 8))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette="viridis")
plt.xlabel('Number of Movies')
plt.ylabel('Genres')
plt.title('Repartition of Genres in the Dataset')
plt.show()
```

```python
genre_counts_only_first = df[df.filter(like='imdb_') == 1].count().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=genre_counts_only_first.values, y=genre_counts_only_first.index, palette="viridis")
plt.xlabel('Number of Movies')
plt.ylabel('Genres')
plt.title('Repartition of Genres in the Dataset (Single Genre Only)')
plt.show()
```

```python
df["genre_count"] = df.filter(regex='^imdb_(?!id$)').count(axis=1)

test = df["genre_count"].value_counts()
print(test)
plt.figure(figsize=(12, 8))
sns.barplot(x=df["genre_count"], y=df["genre_count"].value_counts())
plt.xlabel('Number of genres')
plt.ylabel('Number of movies')
plt.title('Repartition of Genres in the Dataset (Top Genres Only)')
plt.show()
```

```python
genre_cols = [col for col in df_filtered.columns if col.startswith('imdb_')]

# Trouver les genres de priorité 1 et 2
def get_top_two_genres(row):
    sorted_genres = sorted(genre_cols, key=lambda x: row[x])
    return sorted_genres[0], sorted_genres[1]  # Priorité 1 et 2

df_filtered[['genre_1', 'genre_2']] = df_filtered.apply(lambda row: pd.Series(get_top_two_genres(row)), axis=1)
# Création de la matrice de comptage

matrix = pd.crosstab(df_filtered['genre_1'], df_filtered['genre_2'])

# Visualisation
plt.figure(figsize=(10, 6))
sns.heatmap(matrix, annot=True, cmap="coolwarm", fmt="d")
plt.title("Matrice des combinaisons de genres par priorité")
plt.show()
```
