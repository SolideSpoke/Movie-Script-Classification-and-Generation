```python
# Variable
percentage_genre_to_keep = 25
test_size = 0.05
random_state = 42
from collections import Counter
```

```python
df = pd.read_csv("./data/MovieDataThread.csv")
index_drop = df.loc[df['imdb_id'].isnull()].index
df = df.drop(index=index_drop)
df = df.drop(columns=['imdb_id'])
df_filtered = pd.concat([df[['Script','Title']],df.filter(like='imdb_')],axis=1)
df_filtered["genre_count"] = df_filtered.filter(regex='^imdb_(?!id$)').count(axis=1)
df_filtered_one_genre = df_filtered.loc[df_filtered['genre_count'] == 1]
df_filtered_one_genre['filtered_genre'] = df_filtered_one_genre.apply(lambda row: next(iter([col for col in df_filtered_one_genre.columns if "imdb" in col and row[col] == 1.0 ]), 'unknown'), axis=1)
number_genres = len(df_filtered_one_genre['filtered_genre'].unique())
counter = Counter(df_filtered_one_genre['filtered_genre'])
sorted_list = sorted(counter.items(),reverse=True , key=lambda genre : genre[1])[: int(number_genres * (percentage_genre_to_keep / 100))]

genre_to_keep = [genre for genre,_ in sorted_list]
columns_to_keep = ['Script','filtered_genre']

df_movies = df_filtered_one_genre.loc[df_filtered_one_genre['filtered_genre'].isin(genre_to_keep)]
df_movies = df_movies[columns_to_keep].reset_index(drop=True)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
```

```python
def preprocess(script):
    tokens = word_tokenize(script.lower())
    filtered = [token for token in tokens if  token not in stop_words]
    return filtered

X_token = [preprocess(script) for script in df_movies['Script']]
    
```

```python
# Calculer la taille de chaque document en nombre de tokens
document_lengths = [len(tokens) for tokens in X_token]

# Tracer un histogramme des tailles des documents
plt.figure(figsize=(12, 8))
plt.hist(document_lengths, bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Nombre de tokens par script')
plt.ylabel('Nombre de scripts')
plt.title('Distribution de la taille des scripts en nombre de tokens')
plt.show()
```

```python
X_nofiltered_token = [word_tokenize(script) for script in df_movies['Script']]
    
```

```python
# Calculer la taille de chaque document en nombre de tokens
script_lengths = [len(tokens) for tokens in X_token]

# Tracer un histogramme des tailles des documents
plt.figure(figsize=(12, 8))
plt.hist(script_lengths, bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Nombre de tokens par script')
plt.ylabel('Nombre de scripts')
plt.title('Distribution de la taille des scripts en nombre de tokens avec le preprocessing')
plt.show()
```

```python
# Calculer la taille de chaque document en nombre de tokens
script_lengths = [len(tokens) for tokens in X_nofiltered_token]

# Tracer un histogramme des tailles des documents
plt.figure(figsize=(12, 8))
plt.hist(script_lengths, bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Nombre de tokens par script')
plt.ylabel('Nombre de scripts')
plt.title('Distribution de la taille des scripts en nombre de tokens sans le preprocessing')
plt.show()
```

```python
# Nombre total de tokens dans le corpus
total_tokens = sum(len(tokens) for tokens in X_token)
print("Nombre total de tokens dans le corpus :", total_tokens)

# Nombre de tokens uniques (vocabulaire)
unique_tokens = set(token for tokens in X_token for token in tokens)
print("Nombre de tokens uniques (vocabulaire) :", len(unique_tokens))
```

```python

max_length = max(script_lengths)

print("Longueur maximale d’un script :", max_length)
```

```python
from collections import Counter

import matplotlib.pyplot as plt

# Compter les occurrences des tokens
token_counts = Counter(token for tokens in X_token for token in tokens)

# Obtenir les 20 tokens les plus fréquents
most_common_tokens = token_counts.most_common(20)

# Extraire les tokens et leurs fréquences
tokens, frequencies = zip(*most_common_tokens)

# Tracer le graphique
plt.figure(figsize=(12, 8))
sns.barplot(x=frequencies, y=tokens, palette="viridis")
plt.xlabel('Fréquence')
plt.ylabel('Tokens')
plt.title('Fréquence des tokens les plus courants (Top 20)')
plt.show()
```

```python
# Trier les tokens par fréquence décroissante
sorted_token_counts = sorted(token_counts.items(), key=lambda x: x[1], reverse=True)

# Calculer la couverture cumulative
cumulative_coverage = []
cumulative_sum = 0
for token, count in sorted_token_counts:
    cumulative_sum += count
    cumulative_coverage.append(cumulative_sum / total_tokens)

# Trouver combien de tokens couvrent 90 % du corpus
threshold = 0.9
num_tokens_90_percent = next(i for i, coverage in enumerate(cumulative_coverage) if coverage >= threshold) + 1

# Tracer le graphique
plt.figure(figsize=(12, 8))
plt.plot(range(1, len(cumulative_coverage) + 1), cumulative_coverage, label="Couverture cumulative")
plt.axvline(x=num_tokens_90_percent, color='r', linestyle='--', label=f"{num_tokens_90_percent} tokens couvrent 90%")
plt.xlabel("Nombre de tokens")
plt.ylabel("Taux de couverture")
plt.title("Taux de couverture du vocabulaire")
plt.legend()
plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plt

# Calculer la longueur moyenne des scripts pour chaque genre
df_movies['script_length'] = [len(tokens) for tokens in X_token]
average_tokens_per_genre = df_movies.groupby('filtered_genre')['script_length'].mean().sort_values(ascending=False)

# Tracer le graphique
plt.figure(figsize=(12, 8))
average_tokens_per_genre.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Genre')
plt.ylabel('Moyenne du nombre de tokens')
plt.title('Moyenne du nombre de tokens par genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculer le nombre total de stop words dans le corpus
total_stop_words = sum(len([token for token in tokens if token in stop_words]) for tokens in X_nofiltered_token)

# Calculer le ratio
ratio_stop_words = total_stop_words / total_tokens
print(f"Ratio de stop words vs tokens restants : {ratio_stop_words:.2%}")
```

```python
df = pd.read_csv("./data/MovieDataThread.csv")
index_drop = df.loc[df['imdb_id'].isnull()].index
df = df.drop(index=index_drop)
df = df.drop(columns=['imdb_id'])
df_filtered = pd.concat([df[['Script','Title']],df.filter(like='imdb_')],axis=1)

# Liste des genres autorisés
allowed_genres = ['imdb_drama',
 'imdb_comedy',
 'imdb_documentary',
 'imdb_horror',
 'imdb_thriller',
 'imdb_romance']


genre_columns = [col for col in df_filtered.columns if col.startswith("imdb_") and col != "imdb_id"]

df_temp = df_filtered.copy()
df_temp[genre_columns] = df_temp[genre_columns].fillna(0)

df_filtered['genre_list'] = df_temp[genre_columns].apply(
    lambda row: [col for col in genre_columns if row[col] != 0], axis=1
)


df_multilabel = df_filtered[df_filtered['genre_list'].apply(lambda genres: all(g in allowed_genres for g in genres))].reset_index(drop=True)
df_multilabel = df_multilabel[df_multilabel['genre_list'].map(len) > 0].reset_index(drop=True)
```

```python
def preprocess(script):
    tokens = word_tokenize(script.lower())
    filtered = [token for token in tokens if  token not in stop_words]
    return filtered

X_multilabel_token = [preprocess(script) for script in df_multilabel['Script']]
    
```
